#Set in Python
It is used to remove all duplicate values

```python
set1 = set()  #empty set
print(set1)

set2 = set('Ishu Gupta')
print(set2)

set3 = set(['Ishu', 'Gupta', 'Ishu', 'Jamoie', 'Kaila', 'Lima'])
print(set3)
```

```python
print(set3[2]) #indexing can't be accessed in sets
```

```python
#Values in Sets can only be accessed through Loop

for i in set3:
  print(i, end = ' ')
```

# Logical Operators

# Bitwise Operator

```python
a = 10
b = 4

print(a | b) # or
print(a & b) # and
print(~a, ~b) # not
print(a ^ b) # xor
print(a >> 2) # right shift
print(a << 2) # left shift
```

```python
# elif is used for multiple conditions along with if-else

age = int(input("Enter your age"))

if age <=12:
  print("You are a Child")
elif age <=19:
  print("You are a Teenager")
elif age <=39:
  print("You are a Young Adult")
elif age <=60:
  print("You are a legit Adult")
else :
  print("You are an OLDIEE!!")
```

# LOOP CONTROL STATEMENTS

```python
# Break
for i in range(10):
  if i == 7:
    break
  print(i)

```

```python
# Continue
for i in range(10):
  if i ==2:
    continue
  print(i)
```

```python
# Nested Loop (for loop inside a loop)

for i in range(1,4):
  for j in range(1,4):
    print(i,j)
```
